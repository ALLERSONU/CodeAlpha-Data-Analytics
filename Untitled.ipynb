```python
import pandas as pd
from textblob import TextBlob
```

```python
data = {
    "Review": [
        "This book is amazing and very interesting.",
        "I loved reading this book.",
        "The story was boring and too long.",
        "Excellent writing and great characters.",
        "Worst book I have ever read.",
        "It was okay, nothing special.",
        "Fantastic experience, highly recommended.",
        "Not good, I didn't enjoy it.",
        "Average book with decent content.",
        "Absolutely wonderful!"
    ]
}

df = pd.DataFrame(data)

df
```

```python
df["Polarity"] = df["Review"].apply(lambda x: TextBlob(x).sentiment.polarity)

df
```

```python
def get_sentiment(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["Polarity"].apply(get_sentiment)

df
```

```python
print(df["Sentiment"].value_counts())
```

```python
import matplotlib.pyplot as plt

df["Sentiment"].value_counts().plot(kind="bar")

plt.title("Sentiment Analysis")

plt.xlabel("Sentiment")

plt.ylabel("Count")

plt.show()
```

```python
df["Sentiment"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Sentiment Distribution")

plt.ylabel("")

plt.show()
```
